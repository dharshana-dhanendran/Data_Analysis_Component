# Inferential Stat

```python
# Anvoa,hypothesis-compare two values find sililarity
```

```python
from scipy import stats

# Two groups
group1 = [12, 14, 15, 16, 14, 18, 19] #ho-both gropu perform well
group2 = [22, 21, 20, 23, 25, 24, 26]

t_stat, p_val = stats.ttest_ind(group1, group2)
print("t-statistic:", t_stat)
print("p-value:", p_val)

alpha = 0.05  # significance level

if p_val < alpha:
    print("Reject H0: Significant difference between groups")
else:
    print("Fail to reject H0: No significant difference between groups")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table
data = np.array([[10, 20],   # Male
                 [20, 30]])  # Female

# Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof)
print("p-value:", p)
print("Expected frequencies:\n", expected)
```
